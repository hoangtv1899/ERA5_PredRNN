```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
gt = np.load('/scratch/network/hvtran/era5/predrnn-pytorch/checkpoints/mnist_predrnn/test_result/1/gt.npy')
pd = np.load('/scratch/network/hvtran/era5/predrnn-pytorch/checkpoints/mnist_predrnn/test_result/1/pd.npy')
```

```python
print(gt.shape)
print(pd.shape)
```

```python
stepi = 5
var_dicts = {0:'u wind', 1:'v wind', 2:'Sea Surface Temperature',
            3:'Surface Pressure', 4:'Precipitation'}
var = 3

fig, axs = plt.subplots(1,2, figsize=(15,10))
ax0 = axs[0]
im0 = ax0.imshow(gt[0,6+stepi,:,:,var])
divider = make_axes_locatable(ax0)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im0, cax=cax, orientation='vertical')
ax0.set_title('ERA5 '+var_dicts[var])

ax1 = axs[1]
im1 = ax1.imshow(pd[0,stepi,:,:,var])
divider = make_axes_locatable(ax1)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im1, cax=cax, orientation='vertical')
ax1.set_title('PredRNN '+var_dicts[var])
```
